```python
def name_of_game(year,team1,team2,id):
    game_details = games_data.groupby("season").get_group(year).groupby("match_id").get_group(id)
    #Date
    date = game_details["date"].drop_duplicates().tolist()[0]
    print("The Game: {} vs {} on {} id {}".format(team1,team2,date,id))
    innings_1 = games_data.groupby("season").get_group(year).groupby("match_id").get_group(id).groupby("inning").get_group(1)
    print("INNING 1:")
    print("The batting team {} scored {} runs".format(innings_1["batting_team"].iloc[0],innings_1["total_runs"].sum()))
    print("The bowling team {} took {} wickets".format(innings_1["bowling_team"].iloc[0],innings_1["player_dismissed"].count()))
    print()
    players = innings_1["batter"].unique()
    for x in players:
        player_inning = innings_1.groupby("batter")
        
        runs_made = player_inning.get_group(x)["total_runs"].sum()-player_inning.get_group(x)["extra_runs"].sum()
        balls_played = player_inning.get_group(x).shape[0]-player_inning.get_group(x)["extras_type"].value_counts().get("wides",0)
        no_of_fours = player_inning.get_group(x)["batsman_runs"].value_counts().get(4,0)
        no_of_six = player_inning.get_group(x)["batsman_runs"].value_counts().get(6,0)
        strike_rate = (runs_made / balls_played) * 100
        strike_rate = round(strike_rate,2)
        out_bowler = player_inning.get_group(x)["bowler"].iloc[-1]
        out_method = player_inning.get_group(x)["dismissal_kind"].iloc[-1]
        out_by = player_inning.get_group(x)["fielder"].iloc[-1]
        if (out_method == "bowled" or out_method == "lbw" ):
            print("{}:{}:{}:{}:{}:{}   {}::{}".format(x,runs_made,balls_played,no_of_fours,no_of_six,strike_rate,out_method,out_bowler))
        else:
            print("{}:{}:{}:{}:{}:{}   {}::{}".format(x,runs_made,balls_played,no_of_fours,no_of_six,strike_rate,out_method,out_by))


    
    print()
    print()



    
    innings_2 = games_data.groupby("season").get_group(year).groupby("match_id").get_group(id).groupby("inning").get_group(2)
    print("INNING 2:")
    print("The batting team {} scored {} runs".format(innings_2["batting_team"].iloc[0],innings_2["total_runs"].sum()))
    print("The bowling team {} took {} wickets".format(innings_2["bowling_team"].iloc[0],innings_2["player_dismissed"].count()))
    print()

    players = innings_2["batter"].unique()
    for x in players:
        player_inning = innings_2.groupby("batter")
        
        runs_made = player_inning.get_group(x)["total_runs"].sum()-player_inning.get_group(x)["extra_runs"].sum()
        balls_played = player_inning.get_group(x).shape[0]-player_inning.get_group(x)["extras_type"].value_counts().get("wides",0)
        no_of_fours = player_inning.get_group(x)["batsman_runs"].value_counts().get(4,0)
        no_of_six = player_inning.get_group(x)["batsman_runs"].value_counts().get(6,0)
        strike_rate = (runs_made / balls_played) * 100
        strike_rate = round(strike_rate,2)
        out_method = player_inning.get_group(x)["dismissal_kind"].iloc[-1]
        out_bowler = player_inning.get_group(x)["bowler"].iloc[-1]
        out_by = player_inning.get_group(x)["fielder"].iloc[-1]
        if (out_method == "bowled" or out_method == "lbw" ):
            print("{}:{}:{}:{}:{}:{}   {}::{}".format(x,runs_made,balls_played,no_of_fours,no_of_six,strike_rate,out_method,out_bowler))
        else:
            print("{}:{}:{}:{}:{}:{}   {}::{}".format(x,runs_made,balls_played,no_of_fours,no_of_six,strike_rate,out_method,out_by))
            
    
    
    print()
    print()
    

def game_info(year,team1,team2):
    game_of_years = games_data.groupby("season").get_group(year)
    games_of_teams =game_of_years[(game_of_years["batting_team"]==team1)&(game_of_years["bowling_team"]==team2)]["match_id"].drop_duplicates().tolist()
    # for x in games_of_teams:
    name_of_game(year,team1,team2,games_of_teams[-1])
    
game_info("2024","SRH","KKR")
```
